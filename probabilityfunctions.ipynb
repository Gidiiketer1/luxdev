```python
import itertools

def sample_mean(n, data):
    # Calculate population mean (not used further here but included)
    mu = sum(data.values()) / len(data)
    
    # Generate all combinations of keys with sample size n
    combinations = list(itertools.combinations(data.keys(), n))
    
    xbar_list = []
    
    # Calculate sample mean for each combination
    for comb in combinations:
        sample_weights = [data[p] for p in comb]
        sample_mean_value = sum(sample_weights) / n
        xbar_list.append(sample_mean_value)
    
    # Return all combinations and their sample means
    return combinations, xbar_list
```

```python
combinations, xbar_list = sample_mean(sample_size, pumpkin_dict)
for c, m in zip(combinations, xbar_list):
    print(c, m)
```
