### Data wrangling
 involves filtering and sorting data frames, featuring engineering,grouping, andmerging and joining dataframes. 

```python
import pandas as pd
df=pd.read_csv(r"tested.csv")
df.head()
```

```python
df.info()
```

```python
df=df.drop("Cabin",axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
# filtering data
#bolean indexing and comparison operators
adults=df[df["Age"]>=18]
adults.head()
```

```python
df["Age"].mean()
```

```python
adults["Age"].mean()
```

```python
males=df[df["Sex"]=="male"]
males.head()
```

```python
males["Sex"].value_counts()
```

```python
Embarked=df[df["Embarked"]=="Q"]
Embarked.head()
```

```python
Embarked.value_counts()
```

```python
# SORTING DATA
# We use "sort_values" the sort_values method has 2 parameters:
# column ASC (true or false)
df=df.sort_values(by="Fare",ascending=True)
df.head()
```

```python
# Feature engineering involves creating new columns that helps improve your analysis.
df["Age_group"]=pd.cut(df["Age"],bins=[0,12,19,59,100],labels=["Child","Teen","Adult","Senior"])
df.head()
```

```python
#extracting title from name column
df["Title"]=df["Name"].apply(lambda name:name.split(",")[1].split(".")[0].strip())
df.head()
```

```python
df.columns
```

```python
#count of the people who survived in a per class
grouped=df.groupby("Pclass")["Survived"].agg("count")
print(grouped)
```

## merging data

```python
dict1 = {
    "id": [1, 2, 3, 4, 5],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva"],
}
```

```python
dict2 = {
    "id": [1, 2, 3, 4, 5],
    "fruits": ["Apple", "Banana", "Orange", "Mango", "Grapes"],
    }
```

```python
import pandas as pd
df1 = pd.DataFrame(dict1)
df1.head()
```

```python
df2 = pd.DataFrame(dict2)
df2.head()
```

```python
merged=pd.merge(df1,df2,on="id", how="inner")
merged.head()
```
